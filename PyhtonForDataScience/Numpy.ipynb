#### Operations on Axis

```python
# Given a matrix mat of size 3 x 3. Find the maximum numbers from each row , say N1, N2 and N3. 
# Result in a matrix by adding.
mat = np.array([[10,5,9], [2, 20, 6], [8, 3, 30]]).reshape(3,3)
print(mat)
upper_indices = np.mask_indices(3, np.triu, 1)
print(upper_indices)
diag_indices = np.diag_indices(3)
print(diag_indices)
lower_indices = np.mask_indices(3, np.tril, -1)
print(lower_indices)
#get max of Top row
l_max_t = np.max(mat[0])
print(l_max_t)
#get max of Middle row
l_max_m = np.max(mat[1])
print(l_max_m)
#get max of Last row
l_max_l = np.max(mat[2])
print(l_max_l)
mat[upper_indices] = mat[upper_indices] + l_max_t
mat[diag_indices] = mat[diag_indices] + l_max_m
mat[lower_indices] = mat[lower_indices] + l_max_l
print(mat)
```

### Indexing and Selection
1. Cut an image slice (the rocket) out of the original image (the astronaut)
2. On the Red Component of the rocket, all values falling in the range of [100, 150] must be equated to 0.
3. Fix the new rocket image back to its place in the original image.

```python
import os.path
from skimage.io import imread
from skimage import data_dir

img = imread(os.path.join(data_dir, 'astronaut.png'))

import matplotlib.pyplot as plt
plt.imshow(img, cmap = 'Greys') # Pass above used image names in place of 'img' to visualize
```

```python
# 1. Cut an image slice (the rocket) out of the original image (the astronaut)
img_slice = img.copy()
img_slice = img_slice[0:300, 360:480]
plt.figure()
plt.imshow(img_slice, cmap = 'Greys')
```

```python
print(img_slice[:, :, 0])
```

```python
# 2. On the red component of the rocket, all the values are falling in the range of [100, 150] must be equated to 0
img_slice[np.greater_equal(img_slice[:,:,0],100) & np.less_equal(img_slice[:,:,0],150)] = 0
plt.figure()
plt.imshow(img_slice)
```

```python
# 3. Fix the new rocket image back to its place in the original image
img[0:300, 360:480, :] = img_slice
plt.figure()
plt.imshow(img)
```

### Linear Algebra

```python
mat1 = np.arange(4).reshape(2, 2)
mat2 = (np.arange(4)*2).reshape(2, 2)
mat3 = (np.arange(4)*3).reshape(2, 2)

''' Performing multiple dot product in one go'''
print(np.linalg.multi_dot([mat1, mat2, mat3]))
```

```python
a = np.array([[3, 1], [1, 2]])
b = np.array([9, 8])

''' Checking if system of equation has unique solution '''
print(np.linalg.det(a))

''' Since det = 5 which is non-zero. Hence, we can have unique solutions 
Finding unique solution'''
print(np.linalg.solve(a,b))
```

```python
''' Calculating inverse: Since, determitant is non-zero hence matrix in invertible'''
print(np.linalg.inv(a))
```

```python
''' Calculating the Rank of the matrix '''
print(np.linalg.matrix_rank(a))
```

```python
a = np.array([[2, 3, 2], [0, 1, 3], [2, 2, 3]])
b = np.array([1, 2, 3])

print(np.linalg.det(a))
print(np.linalg.solve(a, b))
print(np.linalg.inv(a))
print(np.linalg.matrix_rank(a))
```
